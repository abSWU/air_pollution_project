```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('pollution_data.csv')
```

```python
df
```

```python
pd.set_option('display.max_columns',22) #See all columns
```

```python
df.drop(columns=['Address','County','O3 1st Max Hour', 'CO 1st Max Hour', 'SO2 1st Max Hour', 'NO2 1st Max Hour'], inplace=True) #drop columns I won't use
```

```python
df.dropna(axis = 'index', how = 'any') #drop missing data
```

```python
df.sort_values(by=['O3 Mean', 'CO Mean', 'SO2 Mean', 'NO2 Mean'], ascending = False, inplace=True)
```

```python
df
```

```python
df['TotalPollutants'] = df['O3 Mean'] + df['CO Mean'] + df['SO2 Mean'] + df['NO2 Mean'] 
df.sort_values(by=['TotalPollutants'],ascending=False) 
```

```python
df[['State', 'Date', 'TotalPollutants']]
filt = (df['State'] == 'California')
df.sort_values(['TotalPollutants'], ascending=False)
df[filt]
```

```python
df['AQIMean'] = (df['O3 AQI'] + df['CO AQI'] + df['SO2 AQI'] + df['NO2 AQI'])/4 #Note: This is NOT an accurate measurement for air quality levels
                                                                                #Different pollutants have different AQI formulas and health effects
                                                                                #However, I will continue to use this mean for learning purposes only.
df[['State','Date','AQIMean']].sort_values(by='AQIMean',ascending=False)
```

```python
condensed_df = df[['State','Date','AQIMean']].sort_values(by='AQIMean',ascending=False) #Wanted to see only those 3 columns sorted by highest AQI
```

```python
condensed_df
```

```python
condensed_df['Date'] = pd.to_datetime(condensed_df['Date'], format='%Y-%m-%d') #Convert dates to date time format
```

```python
#Finds avg AQI mean for 2000-2006, will be done 3 more times for the other intervals of time
filt = (condensed_df['Date'] >= '2000') &  (condensed_df['Date'] <= '2005') 
avg_2000 = condensed_df[filt]['AQIMean'].mean() 
avg_2000 #22.96 aqi mean for 2000-2005
```

```python
filt = (condensed_df['Date'] >= '2006') &  (condensed_df['Date'] <= '2011')
avg_2006 = condensed_df[filt]['AQIMean'].mean()
avg_2006 #avg aqi for 2006-2011 was 18.87
```

```python
filt = (condensed_df['Date'] >= '2012') &  (condensed_df['Date'] <= '2017')
avg_2012 = condensed_df[filt]['AQIMean'].mean()
avg_2012 #avg aqi for 2012-2017 was 15.95
```

```python
filt = (condensed_df['Date'] >= '2018')
avg_2018 = condensed_df[filt]['AQIMean'].mean()
avg_2018 #avg aqi for 2018-2023 was 15.21
#The avg AQI seems to have went down as time has gone on (lower AQI = less pollutants)
```

```python
%matplotlib inline
```

```python
#Graph our data - this is for 2000-2005 specifically
filt = (condensed_df['Date'] >= '2000') &  (condensed_df['Date'] <= '2005') 
df2000 = condensed_df[filt][['Date','AQIMean']]
df2000.set_index('Date',inplace=True)
df2000avg=df2000['AQIMean'].resample('ME').mean()
df2000avg.plot(x='Date',y='AQIMean')
```

```python
#Graph our data of avg AQI for our entire DF, sorted by monthly intervals
condensed_df_avg = condensed_df['AQIMean'].resample('ME').mean()
condensed_df_avg.plot(x='Date',y='AQIMean')
```

```python
#Graph our data of avg AQI for our entire DF, sorted by yearly intervals
condensed_df_avg = condensed_df['AQIMean'].resample('YE').mean()
condensed_df_avg.plot(x='Date',y='AQIMean')
```
